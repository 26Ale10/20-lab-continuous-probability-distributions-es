## Distribución Uniforme

Para generar números aleatorios uniformes entre dos valores dados usando scipy, podemos usar el siguiente código o el código que hemos
discutido en clase:

```python
# your code here
# Parámetros de la distribución uniforme
a = 0  # Límite inferior (inclusivo)
b = 10 # Límite superior (inclusivo)
size = 100 # Cantidad de números aleatorios a generar

# Generar números aleatorios
numeros_aleatorios = uniform.rvs(loc=a, scale=b-a, size=size)

plt.hist(numeros_aleatorios, bins=20
         , density=True)
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Uniforme (0, 10)')
plt.show()
```

1. Basándote en el código anterior, escribe una función que genere números aleatorios uniformemente distribuidos. Hay varios requisitos para tu función:
    * Debe aceptar 3 parámetros: 
        * `bottom` - el límite inferior de los números generados
        * `ceiling` - el límite superior de los números generados
        * `count` - cuántos números generar
    * Debe devolver una matriz de números aleatorios distribuidos uniformemente

2. Llama a tu función con 2 conjuntos de parámetros a continuación:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

3. Traza las distribuciones uniformes generadas anteriormente utilizando histogramas, donde el eje x es el valor y el eje y es el recuento. Que el número de intervalos del histograma sea 10.

El resultado debe ser como el siguiente:



```python
# your code here
def Numeros_aleatorios(bottom,ceiling,count):
  aleatoriosSp = uniform.rvs(loc=bottom, scale=ceiling-bottom, size=count)
  aleatoriosNp = np.random.uniform(bottom, ceiling, count)
  return aleatoriosSp,aleatoriosNp

ejemplo1 = Numeros_aleatorios(10,15,100)
ejemplo2 = Numeros_aleatorios(10,60,1000)
```

```python
# your code here
fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(10,5))

axes[0,0].hist(ejemplo1[0], color = 'lightblue', ec =  'black')
axes[0,1].hist(ejemplo2[0], histtype =  'stepfilled', color = 'lightblue', ec = 'black')
axes[1,0].hist(ejemplo1[1], color = 'lightblue', ec =  'black')
axes[1,1].hist(ejemplo2[1], histtype =  'stepfilled', color = 'lightblue', ec = 'black')
axes[0,0].set_yticks(range(0, 120, 20))
axes[0,1].set_yticks(range(0, 120, 20))
axes[1,0].set_yticks(range(0, 120, 20))
axes[1,1].set_yticks(range(0, 120, 20))
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the mean and standard deviation for employee salaries
mean_salary = 60000  # Mean salary in USD
std_deviation = 10000  # Standard deviation in USD

# Generate synthetic data following a normal distribution
num_employees = 1000  # Number of employees
employee_salaries = np.random.normal(mean_salary, std_deviation, num_employees)

# Plot a histogram of employee salaries
plt.figure(figsize=(8, 6))
plt.hist(employee_salaries, bins=30, density=True, alpha=0.6, color='b', label='Employee Salaries')

# Generate a range of x values for the probability density function (PDF)
x_values = np.linspace(min(employee_salaries), max(employee_salaries), 1000)

# Calculate the PDF using the scipy.stats.norm library
pdf_values = norm.pdf(x_values, mean_salary, std_deviation)

# Plot the PDF as a smooth curve
plt.plot(x_values, pdf_values, 'r-', lw=2, label='Normal Distribution PDF')

# Add labels and a legend to the plot
plt.xlabel('Salary (USD)')
plt.ylabel('Probability Density')
plt.title('Employee Salary Distribution')
plt.legend()

# Show the plot
plt.show()
```
