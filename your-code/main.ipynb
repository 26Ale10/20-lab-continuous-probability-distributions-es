## Distribución Uniforme

Para generar números aleatorios uniformes entre dos valores dados usando scipy, podemos usar el siguiente código o el código que hemos
discutido en clase:

```python
# your code here
# Parámetros de la distribución uniforme
a = 0  # Límite inferior (inclusivo)
b = 10 # Límite superior (inclusivo)
size = 100 # Cantidad de números aleatorios a generar

# Generar números aleatorios
numeros_aleatorios = uniform.rvs(loc=a, scale=b-a, size=size)

plt.hist(numeros_aleatorios, bins=20
         , density=True)
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Uniforme (0, 10)')
plt.show()
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
# your code here

def lista_num_exp(muestra,media):
  x=1/media
  return np.random.exponential(scale=1/x, size=muestra)

# Generar 10 números con media 10
muestras10 = lista_num_exp(10, 10)

# Generar 100 números con media 10
muestras100 = lista_num_exp(100, 10)
```

```python
# Graficar las distribuciones
plt.figure(figsize=(10, 6))

# Histograma de la primera muestra (10 números)
plt.subplot(1, 2, 1)
sns.histplot(muestras10, bins=100, alpha=0.7, color='skyblue', label='10 muestras',kde=True)
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Exponencial (10 muestras, Media=10)')
plt.legend()

# Histograma de la segunda muestra (100 números)
plt.subplot(1, 2, 2)
sns.histplot(muestras100, bins=100, alpha=0.7, color='lightcoral', label='100 muestras',kde=True)
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Exponencial (100 muestras, Media=10)')
plt.legend()

plt.tight_layout()
plt.show()
```

¿En qué se diferencian ambas distribuciones?

La media cambia, por lo que la distribución también cambia. 

## Distribución Exponencial de datos reales

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco? 

Escribe un código en python para resolver este problema

```python
# your code here
from scipy.stats import expon

lambda_param = 1/10  # Media = 10 minutos, por lo tanto lambda = 1/media

tiempo = 15
probabilidad = expon.cdf(tiempo, scale=1/lambda_param)
```

¿Cuál es la probabilidad de que el cliente pase más de 15 minutos?

```python
# your code here

print(f"La probabilidad de que un cliente pase menos de 15 minutos en el banco es: {probabilidad:.3f}")
```
